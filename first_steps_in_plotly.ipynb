```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np
import pandas as pd
import scipy.stats as spst

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (18,12)
%matplotlib inline

iplot([{"x": [1, 2, 3], "y": [3, 1, 6]}])
```

```python
# make a scatter plot using matplotlib first
x = np.random.uniform(low=0, high=1, size=100)
eps = 0.1 * np.random.randn(100)
y = 0.4*x+0.5 + eps

fig1 = plt.figure(1)
ax1 = fig1.add_subplot(1,1,1)
ax1.scatter(x, y)
```

```python
# making analogous scatter plot using plotly to compare the quality
sc1 = go.Scatter(
    x = x,
    y = y,
    mode = "markers"
)
iplot([sc1], filename="First plotly scatter plot")
```
